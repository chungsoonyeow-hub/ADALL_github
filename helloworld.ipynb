```python
#### addd a comment to test !!!!!!

import pandas as pd
import matplotlib.pyplot as plt

# Example: Replace this with the raw URL of your GitHub file
github_raw_url = 'https://raw.githubusercontent.com/chungsoonyeow-hub/ADALL_github/refs/heads/main/laptop_prices_2024_sgd_TL.csv'
try:
    df = pd.read_csv(github_raw_url)
    print("Successfully loaded data from GitHub!")
    display(df.head())
    #plot each df col graph
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].astype('category')
            df[col] = df[col].cat.codes
            df[col].plot(kind='hist', title=col)
            plt.show()
        else:
            df[col].plot(kind='hist', title=col)
            plt.show()

    #plot each df col that is numeric with histogram
    for col in df.columns:
        if df[col].dtype != 'object':
            df[col].plot(kind='hist', title=col)
            plt.show()


except Exception as e:
    print(f"Error loading data: {e}")
    print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")

# #plot each df col graph
# for





# for col in df.columns:
#     if df[col].dtype == 'object':
#         df[col] = df[col].astype('category')
#         df[col] = df[col].cat.codes
#         df[col].plot(kind='hist', title=col)
#         plt.show()
#     else:
#         df[col].plot(kind='hist', title=col)
#         plt.show()

```
